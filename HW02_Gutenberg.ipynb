# Hyponyms
- Which one of those 10 words has the largest number of hyponyms? 
- List all hyponyms of those 10 most frequently used long words.

```python
def get_hyponyms(synset):
    hyponyms = set()
    for hyponym in synset.hyponyms():
        hyponyms |= set(get_hyponyms(hyponym))
    return hyponyms | set(synset.hyponyms())
```

```python
gov = wn.synsets('government', 'n')[0]
hypos = lambda s:s.hyponyms()
print(list(gov.closure(hypos)))
```

```python
# An example to see the hyponyms of "american.n.01"
government = wn.synset('government.n.01')   
types_of_GOV = list(set([w for s in government.closure
                         (lambda s:s.hyponyms()) 
                         for w in s.lemma_names()]))
types_of_GOV
```

```python
def size(s): 
    layer = [s] 
    total = 0 
    while layer:
        total += len(layer)
        layer = [h for c in layer for h in c.hyponyms()] 
    return total
```
